```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
xls = pd.ExcelFile('collections_list.xlsx')
df1 = pd.read_excel(xls, 'main')
```

```python
df1
```

```python
collection_list = [x for x in df1['Name']]
```

```python
collection_list
```

```python
collection_list_two = []
for i in collection_list:
  collection_list_two.append(i.replace(" ", "-"))

  
```

```python
collection_list_two
```

```python
collection_list_three = []
for i in collection_list_two:
  collection_list_three.append(i.lower())
```

```python
collection_list_three
```

```python
df_uno = pd.DataFrame()

#collections = ["cryptopunks", "art-blocks", "boredapeyachtclub", "0n1-force","888-inner-circle", 'anonymice']

for collection in collection_list_three:
  
  url = "https://api.opensea.io/api/v1/assets?order_by=sale_count&order_direction=desc&offset=0&limit=50&collection=" + collection

  JSONContent = requests.get(url).json()

  content = json.dumps(JSONContent, indent = 2, sort_keys=True)

  JSONContent.keys()

  df_uno = df_uno.append(JSONContent["assets"])

```
